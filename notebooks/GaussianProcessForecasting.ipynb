Gaussian Processes (GP) have been used for time series forecasting.
In this context, the GPs use, as covariate, only the time variable *t*.
$$f \sim \text{GP}(\mu(t), \sigma(t)^2) + \epsilon_t$$

Where $\mu(t)$ is the mean function, $\sigma(t)^2$ is the variance function and $\epsilon_t \sim \mathcal{N}(0,\sigma^2)$. The variance function is defined through the specification of some *kernels* $k(x,x')$.

The use of GPs in the context of time series leads to some benefits but also to a major drawback.

**Benefits:**
- the time series *does not* necessarily have to be stationary, as the GP is - at least in theory - able to fit arbitrary complex trends. 
- the appropriate choice of kernels and their parameters' tuning via the maximization of the marginal log-likelihood allows the model to learn the seasonality automatically.

**Drawback:**
- GPs are extremely flexible models, prone to overfitting the training set, and their marginal log-likelihood may be very difficult to optimize. It is essential that an appropriate maximization algorithm is chosen as the commonly used Nelder-Mead and L-BFGS may fail, since the former does not deal well with multivariate optimization, and the latter requires the computation of gradients, which may be quite unstable when the likelihood is too *wiggly*. For this reason, we use a Differential Evolution algorithm, which is much more stable and less likely to get trapped into local minima.

```python
# first, we import all the libraries.
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import DotProduct, WhiteKernel, \
                                             ConstantKernel, ExpSineSquared, Matern
import os
os.makedirs("GaussianProcessForecasting", exist_ok=True)
plt.style.use("bmh")
plt.rcParams["figure.dpi"] = 150
plt.rcParams["figure.figsize"] = (15,5)

# read data set
df = pd.read_csv("../data/airpassengers_set.csv", parse_dates=True, index_col=0)
# add column of time stamps (as floats)
df["t"] = np.arange(df.shape[0])
df.head()
```

```python
df[["Passengers"]].plot()
```

The next step is splitting the data into training set and test set. We will use the last two years of data for testing.

```python
Ntest = 24
train, test = df.iloc[:-Ntest,:], df.iloc[-Ntest:, :]
train_idx = df.index <= train.index[-1]
test_idx = ~train_idx
X_train = train["t"].values[:,None]
y_train = train["Passengers"].values
X_test = test["t"].values[:,None]
y_test = test["Passengers"].values

df["train"] = train["Passengers"]
df["test"] = test["Passengers"]
df.drop("Passengers", axis=1, inplace=True)
df[["train","test"]].plot()
```

```python
df.head()
```

Great! Now all the data is split into training and test. We are now ready to instantiate a GP.

```python
# instantiate a simple kernel
kernel = DotProduct()*ConstantKernel() + WhiteKernel()
# instantiate and fit model
gp_model = GaussianProcessRegressor(kernel = kernel, random_state=2023)
gp_model.fit(X_train, y_train)

# get in-sample prediction
df.loc[train_idx, "train_predicted"] = gp_model.predict(X_train)
# get out-of-sample prediction
df.loc[test_idx, "test_predicted"] = gp_model.predict(X_test)

# plot the results
plt.plot(df["train"], color="k", lw=1.0, label="Training, Ground Truth")
plt.plot(df["train_predicted"], color="k", lw=2.5, ls="dotted", label="Training, Prediction")
plt.plot(df["test"], color="r", lw=1.0, label="Test, Ground Truth")
plt.plot(df["test_predicted"], color="r", lw=2.5, ls="dotted", label="Test, Prediction")
plt.xlim([df.index[0],df.index[-1]])
plt.title("Gaussian Process Prediction for Dot Product Kernel");
plt.legend();
```

The results of the Gaussian Process are extremely underwhelming. We only get a trend and no seasonality. We can do better than this by including other kernels which are more suitable for a periodic time series.

```python
# instantiate a simple kernel
kernel = ExpSineSquared() * ( DotProduct()*ConstantKernel() + WhiteKernel())
# instantiate and fit model
gp_model = GaussianProcessRegressor(kernel = kernel, random_state=2023)
gp_model.fit(X_train, y_train)

# get in-sample prediction
df.loc[train_idx, "train_predicted"] = gp_model.predict(X_train)
# get out-of-sample prediction
df.loc[test_idx, "test_predicted"] = gp_model.predict(X_test)

# plot the results
plt.plot(df["train"], color="k", lw=1.0, label="Training, Ground Truth")
plt.plot(df["train_predicted"], color="k", lw=2.5, ls="dotted", label="Training, Prediction")
plt.plot(df["test"], color="r", lw=1.0, label="Test, Ground Truth")
plt.plot(df["test_predicted"], color="r", lw=2.5, ls="dotted", label="Test, Prediction")
plt.xlim([df.index[0],df.index[-1]])
plt.title("Gaussian Process Prediction for Dot Product Kernel and ExpSinSquared Kernel");
plt.legend();
```

Out of the box, the fit is underwhelming for several reasons:
1) The default L-BFGS optimizer does not correctly minimize the marginal log-likelihood. Thus, the parameters of the covariance matrix are not a good fit for the data.
2) The range of the output variable might be too large. We might need normalization.
We will first try and work on 1) and then proceed with 2), to then combine both the approaches and observe the result.

## 1) Inclusion of a custom optimizer.

The `GaussianProcessRegressor` class allows the definition of a custom optimizer. In this case, we choose the powerful `DifferentialEvolution` optimizer from the scipy package.

```python
from scipy.optimize import differential_evolution

def customOptimizer(obj_func, initial_theta, bounds):
    # * 'obj_func': the objective function to be minimized, which
    #   takes the hyperparameters theta as a parameter and an
    #   optional flag eval_gradient, which determines if the
    #   gradient is returned additionally to the function value
    # * 'initial_theta': the initial value for theta, which can be
    #   used by local optimizers
    # * 'bounds': the bounds on the values of theta
    # Returned are the best found hyperparameters theta and
    # the corresponding value of the target function.
    
    # careful! obj_func returns also an array, in addition to the function value
    np.random.seed(2023)
    wrapper_obj_func = lambda x : obj_func(x)[0]
    min_val = (differential_evolution(wrapper_obj_func, bounds=bounds)).x
    return min_val, wrapper_obj_func(min_val)


# instantiate a simple kernel
kernel = DotProduct()*ConstantKernel() + WhiteKernel()
# instantiate and fit model
gp_model = GaussianProcessRegressor(kernel = kernel, optimizer=customOptimizer, random_state=2023)
gp_model.fit(X_train, y_train)

# get in-sample prediction
df.loc[train_idx, "train_predicted"] = gp_model.predict(X_train)
# get out-of-sample prediction
df.loc[test_idx, "test_predicted"] = gp_model.predict(X_test)

# plot the results
plt.plot(df["train"], color="k", lw=1.0, label="Training, Ground Truth")
plt.plot(df["train_predicted"], color="k", lw=2.5, ls="dotted", label="Training, Prediction")
plt.plot(df["test"], color="r", lw=1.0, label="Test, Ground Truth")
plt.plot(df["test_predicted"], color="r", lw=2.5, ls="dotted", label="Test, Prediction")
plt.xlim([df.index[0],df.index[-1]])
plt.title("Gaussian Process Prediction for Dot Product Kernel, with Differential Evolution");
plt.legend();
```

Well, this did not lead to any benefits. How about the model including the seasonality?

```python
# instantiate a simple kernel
kernel = ExpSineSquared() * ( DotProduct()*ConstantKernel() + WhiteKernel())

# instantiate and fit model
gp_model = GaussianProcessRegressor(kernel = kernel, random_state=2023, optimizer=customOptimizer)
gp_model.fit(X_train, y_train)

# get in-sample prediction
df.loc[train_idx, "train_predicted"] = gp_model.predict(X_train)
# get out-of-sample prediction
df.loc[test_idx, "test_predicted"] = gp_model.predict(X_test)

# plot the results
plt.plot(df["train"], color="k", lw=1.0, label="Training, Ground Truth")
plt.plot(df["train_predicted"], color="k", lw=2.5, ls="dotted", label="Training, Prediction")
plt.plot(df["test"], color="r", lw=1.0, label="Test, Ground Truth")
plt.plot(df["test_predicted"], color="r", lw=2.5, ls="dotted", label="Test, Prediction")
plt.xlim([df.index[0],df.index[-1]])
plt.title("Gaussian Process Prediction for Dot Product Kernel and ExpSinSquared Kernel, with Differential Evolution");
plt.tight_layout()
plt.savefig("./GaussianProcessForecasting/fig1.png")
plt.legend();
```

**Results:**
- Gaussian Processes *can* be used for forecasting.
- Normalization does not seem to bring benefits to the fit.
- L-BFGS is unable to find the global minimum of the marginal log-likelihood. More involved optimization methods, such as the evolutionary Differential Evolution algorithm should be used in these circumstances (although fitting the model ends up taking longer).


## Sampling from the posterior distribution

It is interesting to see what the samples from the posterior distribution look like.

```python
# instantiate a simple kernel
kernel = ExpSineSquared() * ( DotProduct()*ConstantKernel() + WhiteKernel())

# instantiate and fit model
gp_model = GaussianProcessRegressor(kernel = kernel, random_state=2023, optimizer=customOptimizer)
gp_model.fit(X_train, y_train)

# get 100 samples from the posterior distribution
posterior_samples = gp_model.sample_y(df["t"].values[:,None], n_samples=100, random_state=2023)

# plot the 100 samples from the posterior
plt.plot(df.index, posterior_samples, alpha=0.01, color="gray");
df[["train", "test"]].plot(ax=plt.gca())
plt.xlabel("Month");
plt.ylabel("Passengers")
plt.title("Samples from the Posterior")
plt.xlim([df.index[0],df.index[-1]])
plt.tight_layout()
plt.savefig("./GaussianProcessForecasting/fig2.png")
```

We see that the samples from the posterior all tend to be close to each other, which reinforces our belief in the appropriateness of the kernel specification.
